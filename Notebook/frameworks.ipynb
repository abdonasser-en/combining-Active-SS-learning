# Overview on Framework

### Framework 1 :

Initially, the framework trains a model on the small set of labeled data. Then, an iteration begins for a fixed number of rounds, denoted by n. In each round, the framework operates differently based on whether the iteration is even or not.

### Framework 2

Similar to Framework 1, the change here involves measuring the stabiliity of the model on unlabeled data rather than relying on whether the iteration is even or not.
